```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/13_dropout_layer/sonar_dataset.csv", header=None)

df[60] = pd.get_dummies(df[60], drop_first=True)
```

```python
X, y = df.drop(columns=60), df[60]
kfold = StratifiedKFold(n_splits=4, shuffle=True)
```

Apply Kfold to non dropout regularized neural networkb

```python
scores = {"no_dropout": []}

for train_i, test_i in kfold.split(X, y):
    model = keras.Sequential([
        keras.layers.Dense(120, input_shape=(60,), activation="relu"),
        #keras.layers.Dropout(0.8),
        keras.layers.Dense(300, activation="relu"),
        #keras.layers.Dropout(0.8),
        keras.layers.Dense(300, activation="relu"),
        keras.layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer = "adam",
        loss = "binary_crossentropy",
        metrics = ["accuracy",]
    )
    model.fit(X.loc[train_i], y.loc[train_i], epochs=300)
    scores["no_dropout"].append((model.evaluate(X.loc[test_i], y.loc[test_i])[1]))
scores["no_dropout"] = np.mean(scores["no_dropout"])
```

Apply Kfold to dropout regularized neural network

```python
scores["dropout"] = []

for train_i, test_i in kfold.split(X, y):
    model = keras.Sequential([
        keras.layers.Dense(120, input_shape=(60,), activation="relu"),
        keras.layers.Dropout(0.7),
        keras.layers.Dense(300, activation="relu"),
        keras.layers.Dropout(0.7),
        keras.layers.Dense(300, activation="relu"),
        keras.layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer = "adam",
        loss = "binary_crossentropy",
        metrics = ["accuracy",]
    )
    model.fit(X.loc[train_i], y.loc[train_i], epochs=300)
    scores["dropout"].append((model.evaluate(X.loc[test_i], y.loc[test_i])[1]))
scores["dropout"] = np.mean(scores["dropout"])
```

70% Dropout with 4 folds gave the max performance with a mean acc of 0.8701923042535782
